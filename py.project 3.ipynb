```python
# Welcome message
print("=========|WELCOME TO STUDENT DATA ORGANISATION|=========")
print("This program helps you manage student records efficiently.")

student = []           # list
subject_set = set()    # set

while True:
    print("\n======|MENU|======")
    print("1. Add Student")
    print("2. Display All Students")
    print("3. Update Student")
    print("4. Delete Student ")
    print("5. Display Unique Subject")
    print("6. Exit")

    choice = input("Enter your choice: ")

    # Add Student
    if choice == "1":
        name = input("Enter Name:")
        age = int(input("Enter Age:"))
        grade = input("Enter Grade:")
        sid = input("Enter Student ID:")
        dob = input("Enter Date of Birth:")
        sub = input("Enter Subjects (comma-seprated):")

        subject_list = [s.strip() for s in sub.split(",")]
        subject_set.update(subject_list)

        student_info = (sid,dob)

        student_data= {
            "id": sid,
            "info": student_info,
            "name": name,
            "age": age,
            "grade": grade,
            "subject": subject_list
        }

        student.append(student_data)


        print("\nStudent Added Successfully!")

        # String Formatting
        print(f"Name: {name}, Age: {age}")
        print("Grade:{} | ID: {}".format(grade, sid))
        print("DOB: %s" % dob)

    # Display All Students
    elif choice == "2":
        if len(student) == 0:
            print("No student records found.")

        else:
            print("\n=======STUDENT RECORDS=======")
            for s in student:
                print("-----------------------------------------")
                print(f"""                                    
                    Student ID : {s["id"]}
                    DOB        : {s["info"][1]}
                    Name       : {s["name"]}
                    Age        : {s["age"]}
                    Grade      : {s["grade"]}
                    Subjects   : {",".join(s["subject"])}""")
                print("------------------------------------------")

    # Update Student
    elif choice == "3":
        sid = input("Enter Student ID to Update: ")
        found = False

        for s in student:
            if s["id"] == sid:
                found = True
                s["name"] = input("enter New name:")
                s["age"] = int(input("Enter New Age:"))
                s["grade"] = input("Enter New Grade:")

                new_subject = input("Enter New Subjects (comma seprated):").split(",")
                s["subject"] = [i.strip() for i in new_subject]
                subject_set.update(s["subject"])
                print("Student Updated Successfully!")
                break

                if not found:
                    print("Student ID not found.")

    # Delete Student
    elif choice == "4":
        sid = input("Enter Student ID to Delete: ")
        found = False

        for i in range(len(student)):
            if student[i]["id"] == sid:
                del student[i]
                found = True
                print("Student ID deleted successfully.")
                break
            
            if not found:
                print("Student ID not found.")

    # Display Unique Subject
    elif choice == "5":
        if len(subject_set) == 0:
            print("No subject available.")
        else:
            print("\nUnique Subjects:")
            for sub in sorted(subject_set):
                print(sub)

    # Exit
    elif choice == "6":
        print("\nThank you for using the Student Data Organisation!")
        print("Program Ended Successfully.")
        break
    else:
        print("Invalid Choice!\nPlease try again.")
```
